```python
extra = pd.DataFrame({'시도교육청' : ['서울','서울','서울','인천','인천'], 
                      '폐교명' : ['성수공업고등학교', '도봉고등학교','화양초등학교','동암초등학교','신촌초등학교'],
                      '폐교연도' : [2024,0,0,0,0],
                      '지역교육청' : ['미정','미정','미정','미정','미정'],
                      '급별' : ['고','고','초','초','초'],
                      '활용현황' : ['미활용','미활용','미활용','미활용','미활용'],
                      '주소' : ['서울 성동구 뚝섬로 365 성수공업고등학교','서울 도봉구 도봉산길 27 도봉고등학교','서울 광진구 군자로 9','인천 부평구 경원대로 1215','인천 부평구 동암남로35번길 4']})
extra

list = list.append(extra, ignore_index = True)
```

```python
tmp = list[list['활용현황'] != '미활용'].index
list.drop(tmp,axis='index',inplace=True)
list['폐교명'] = list['폐교명'].apply(lambda x: re.sub(r'\([^)]*\)', '',x) if '(' in x else x)
last = []
for i in list['폐교명'] :
  if i.strip()[-1] == '중' :
    i += '학교'
    last.append(i)
  elif i.strip()[-1] == '초' or i.strip()[-1] == '고' :
    i += '등학교'
    last.append(i)
  else :
    last.append(i)

list['폐교명'] = last
```

```python
####### 도로명주소 위도 경도 값으로 바꿔주기 ########
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')
# 위도, 경도 반환하는 함수
def geocoding(address):
    geo = geo_local.geocode(address)
    x_y = [geo.latitude, geo.longitude]
    return x_y

def find_pos(name,addr) :
  headers={
      "Authorization" : "KakaoAK da91f68bb8a4dc7fe01d213cb54e6026"
  }
  try :
    url = f'https://dapi.kakao.com/v2/local/search/keyword.json?query={name}'
    getJson = requests.get(url,headers=headers).json()['documents'][0]
  except IndexError :
    url = f'https://dapi.kakao.com/v2/local/search/keyword.json?query={addr}'
    getJson = requests.get(url,headers=headers).json()['documents'][0]
  y = float(getJson['x'])
  x = float(getJson['y'])
  x_y = [x, y]
  return x_y


#####주소를 위,경도 값으로 변환하기 #####
latitude = []
longitude =[]

for name,addr in zip(list['폐교명'],list['주소']) :
  try :
    x,y = geocoding(addr)
  except AttributeError :
    try :
      x,y = geocoding(name) 
    except AttributeError :
      try :
        x,y = find_pos(name,addr)
      except AttributeError :
        x,y = 0,0
      except IndexError :
        x,y = 0,0
  latitude.append(x)
  longitude.append(y)
list['위도'] = latitude
list['경도'] = longitude
```

```python
tmp = list[list['위도'] == 0].index
list.drop(tmp,axis='index',inplace=True)
```

```python
list['위도'] = list['위도'].astype(float)
list['경도'] = list['경도'].astype(float)
```

```python
서울 = list[list['시도교육청']=='서울']
인천 = list[list['시도교육청']=='인천']
경기 = list[list['시도교육청']=='경기']
```

```python
df = pd.concat([서울,인천,경기],ignore_index=True)
```
